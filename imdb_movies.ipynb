```python
#comparing the relation between how long the movie and how much it is profitable
lng_prf = []
mean = df['Revenue (Millions)'].quantile(.50)
for index , rows in df.iterrows() :
    if (rows['Revenue (Millions)'] >= mean ) & (rows['Runtime (Minutes)'] >= 120 ) :
        lng_prf.append('lng_prfbl')
    elif (rows['Revenue (Millions)'] >= mean) & (rows['Runtime (Minutes)'] < 120 ) :
        lng_prf.append('shrt_prfbl')
    elif (rows['Revenue (Millions)'] < mean) & (rows['Runtime (Minutes)'] < 120 ) :
        lng_prf.append('shrt_notprfbl')  
    elif (rows['Revenue (Millions)'] < mean) & (rows['Runtime (Minutes)'] >= 120 ) :
        lng_prf.append('lng_notprfbl')
```

```python
df['lng_prf']=lng_prf
```

```python
df
```

```python
df['lng_prf'].value_counts()
#results show that long movies are less but can be profitable more than short movies
```
